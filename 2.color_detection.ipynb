YELLOW COLOR OBJECT DETECTION

```python
import cv2
import numpy as np

img = cv2.imread('colors.jpeg')
cv2.imshow('Orginal Image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
##convert to hsv image

hsv_image = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
cv2.imshow('hsv Image',hsv_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
lower_yellow = np.array([20, 100, 100])
upper_yellow = np.array([30, 255, 255])

mask = cv2.inRange(hsv_image,lower_yellow,upper_yellow)
cv2.imshow('mask Image',mask)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
x,y,w,h = cv2.boundingRect(mask)
cv2.rectangle(img,(x,y),(x+w,x+h),(255,0,0),1)
cv2.imshow('res Image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

TRAFFIC SIGNAL DETECTION

```python
import cv2
import numpy as np

# Load the image
img = cv2.imread("traffic_light.jpeg")
img = cv2.resize(img, (500, 700))
hsv_image = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)

# HSV color ranges
color_ranges = {
    "Red": [(0, 120, 70), (10, 255, 255)],
    "Yellow": [(20, 100, 100), (40, 255, 255)],
    "Green": [(40, 50, 50), (90, 255, 255)],
}

for color , (lower,upper) in color_ranges.items():
    lower  = np.array(lower,dtype="uint8")
    upper  = np.array(upper,dtype="uint8")

    mask = cv2.inRange(hsv_image,lower,upper)

    ##find countours
    contours , heirarchy = cv2.findContours(mask,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
    for contour in contours:
        # print(cv2.contourArea(contour))
        if cv2.contourArea(contour) > 500:
            print(cv2.contourArea(contour))
            x,y,w,h = cv2.boundingRect(contour)
            cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),3)
            cv2.putText(img,f"{color} Light",(x,y-10),cv2.FONT_HERSHEY_SIMPLEX,1,(255,0,0),3)
cv2.imshow("final",img)
cv2.waitKey(0)
cv2.destroyAllWindows()
print(cv2.contourArea(contour))
```
