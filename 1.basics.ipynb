```python
# BLURRING THE IMAGES


# Blur Type	Use Case

# Averaging  	Basic noise reduction
# Gaussian	    Smoother, used in object detection            Used in object detection (YOLO, Faster R-CNN) preprocessing.
# Median	    Best for noise removal without losing edges  used for fingerprint recognition, face smoothing       
# Bilateral	    Best for edge-preserving blur                Used in cartooning, facial smoothing, and HDR processing.
```

```python
### blurring the image
image = cv2.imread('man.jpg')
plt.imshow(image)
plt.axis('off')
plt.show()
print(image.shape)

##blur image

# 1.averaging

blur_image = cv2.blur(image,(5,5))  ## cv2.blur(image,kernel_size) more the kernel size more the blur of image
plt.imshow(blur_image)
plt.axis('off')
plt.show()

# 2.guassian blur 

blur_image = cv2.GaussianBlur(image,(5,5),0)  ## cv2.blur(image,kernel_size(usually odd 3,5,7,..),standard_deviation(if 0 it automatically calulates based on kernel size)) more the kernel size more the blur of image
plt.imshow(blur_image)
plt.axis('off')
plt.show()

## 3.median blur
blur_image = cv2.medianBlur(image,5)  ## cv2.blur(image,kernel_size of 1dim) more the kernel size more the blur of image
plt.imshow(blur_image)
plt.axis('off')
plt.show()

##4.bilateral blur

blur_image = cv2.blur(image,(5,5))  ## cv2.blur(image,kernel_size) more the kernel size more the blur of image
plt.imshow(blur_image)
plt.axis('off')
plt.show()
```

THRESHOLD

```python
##threshold
# ✅ Thresholding in OpenCV
# Thresholding is a technique used to convert an image into a binary image (black & white) by applying a threshold value.
# It is commonly used in image segmentation and feature extraction.
```

```python
image = cv2.imread('parrot.jpeg')
plt.imshow(image)
plt.axis('off')
plt.show()

# 1.binary threshold

ret,binary = cv2.threshold(image,127,255,cv2.THRESH_BINARY)  ##cv2.threshold(image,127(values above 127 are set to white(255) and below 127 are set to black(0)),255(max pixel val),threshold type)
plt.imshow(binary)
plt.axis('off')
plt.show()

# 2. inverted binary 
res,inverted_binary = cv2.threshold(image,127,255,cv2.THRESH_BINARY_INV)
plt.imshow(inverted_binary)
plt.axis('off')
plt.show()

##3.truncate threshold

##above 127 is set to 127 nd below 127 are remain unchanged
res,truncate = cv2.threshold(image,127,255,cv2.THRESH_TRUNC)
plt.imshow(truncate)
plt.axis('off')
plt.show()

##4. toZero
##below 127 is set to 0 nd above 127 are remain unchanged
res,toZero = cv2.threshold(image,127,255,cv2.THRESH_TOZERO)
plt.imshow(toZero)
plt.axis('off')
plt.show()

##5.toZeroInverse
##above 127 is set to 0 nd below 127 are remain unchanged
_, to_zero_inv = cv2.threshold(image, 127, 255, cv2.THRESH_TOZERO_INV)
plt.imshow(to_zero_inv)
plt.axis('off')
plt.show()

# Adaptive Thresholding (for Varying Lighting Conditions)
# If an image has uneven lighting, simple thresholding won’t work well. Use adaptive thresholding instead:

# Parameters:
# 255 → Maximum pixel value.
# cv2.ADAPTIVE_THRESH_GAUSSIAN_C → Uses a weighted sum of neighboring pixels.
# cv2.THRESH_BINARY → Type of thresholding.
# 11 → Block size (region size for calculating threshold).
# 2 → Constant subtracted from the mean.

uneven_image = cv2.imread('uneven.jpg')
grayscale_uneven_image = cv2.cvtColor(uneven_image,cv2.COLOR_BGR2GRAY)
plt.imshow(grayscale_uneven_image)
plt.axis('off')
plt.show()

adaptive = cv2.adaptiveThreshold(grayscale_uneven_image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)
plt.imshow(adaptive)
plt.axis('off')
plt.show()


```

```python
# # types 
# Sobel: Detects edges in specific directions (X & Y).
# Laplacian: Detects edges in all directions but is sensitive to noise.
# Canny: Best method for accurate edge detection using gradient analysis.
```

```python
# Canny Edge Detection is the most accurate and widely used method in real-time applications like self-driving cars, facial recognition, and medical imaging.
# cv2.Canny(image, 100, 200)
# 100: Lower threshold – weak edges below this are removed.
# 200: Upper threshold – strong edges are retained.
```

CONTOURS IN OPENCV


```python
##steps

## 1.Convert the image to grayscale.
# 2. Apply thresholding or Canny edge detection.
# 3.Use cv2.findContours() to extract contours.
# 4.Draw the detected contours using cv2.drawContours().
```

```python
##bsics in countours

import cv2
import numpy as np

img = cv2.imread('birds.png')
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
ret,thresh = cv2.threshold(gray,127,255,cv2.THRESH_BINARY)
contours , hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
cv2.drawContours(img,contours,-1,(0,255,0),1)
# Show results
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np

##read image
img = cv2.imread('birds.png')

##convert to grayscale
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

##apply threshold
ret,thresh = cv2.threshold(gray,127,255,cv2.THRESH_BINARY)


##findcountors
contours , hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)

##draw countors
cv2.drawContours(img,contours,-1,(0,255,0),1)

### draw box for each contour
count = 0 ## count no.of objects
for contour in contours:
    if cv2.contourArea(contour) > 40:
        x , y , w , h = cv2.boundingRect(contour)  ## return top-left coordinatinators(x,y) , weight w and height h
        cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),1)
        count = count+1
plt.imshow(img)
plt.axis('off')
plt.show()

print('no.of objects in the image are : ',count)
```

```python
import cv2
import numpy as np

# Load the image
img = cv2.imread('birds.png')  # Replace with your image path

plt.title('Original image')
plt.imshow(img)
plt.axis('off')
plt.show()

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  # Convert to grayscale
plt.title('Gray Scale image')
plt.imshow(gray)
plt.axis('off')
plt.show()

# Apply thresholding to get a binary image
_, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

# Find contours
contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

# Draw contours on the original image
cv2.drawContours(img, contours, -1, (0, 255, 0),1)  # Green color, thickness 2

# Show results
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
import cv2

cap = cv2.VideoCapture(0)  # Open webcam

while True:
    ret, frame = cap.read()
    if not ret:
        break

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    edges = cv2.Canny(gray, 50, 150)  # Detect edges
    contours, _ = cv2.findContours(edges, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
    cv2.drawContours(frame, contours, -1, (0, 255, 0), 2)  # Green contours

    cv2.imshow("Live Contours", frame)

    if cv2.waitKey(1) & 0xFF == 27:  # Press ESC to exit
        break

cap.release()
cv2.destroyAllWindows()
```
