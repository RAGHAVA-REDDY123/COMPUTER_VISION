```python
import cv2
import numpy 
import mediapipe as mp
import time
```

```python
print(10)
```

```python
cap = cv2.VideoCapture(0)

mp_hands = mp.solutions.hands
hands = mp_hands.Hands(min_detection_confidence=0.5,min_tracking_confidence=0.5,max_num_hands=2)
mp_draw = mp.solutions.drawing_utils
while True:
    success,image = cap.read()

    if not success:
        break

    RGB_image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

    result = hands.process(RGB_image)

    if result.multi_hand_landmarks:
        for hand_landmarks in result.multi_hand_landmarks:
            mp_draw.draw_landmarks(image,hand_landmarks,mp_hands.HAND_CONNECTIONS)
            print(result.multi_hand_landmarks)

    cv2.imshow('Hands_Detected',image)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)

mp_hands = mp.solutions.hands
hands = mp_hands.Hands(min_detection_confidence=0.5,min_tracking_confidence=0.5,max_num_hands=4)
mp_draw = mp.solutions.drawing_utils
while True:
    success,image = cap.read()

    if not success:
        break

    RGB_image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)

    result = hands.process(RGB_image)

    if result.multi_hand_landmarks:
        for hand_landmarks in result.multi_hand_landmarks:
            for id,lm in enumerate(hand_landmarks.landmark):

                # print(id,lm)

                ##convert into pixels
                h,w,c = image.shape
                cx ,cy = int(lm.x*w),int(lm.y*h)

                print(id,cx,cy)

                if id==4:
                    cv2.circle(image,(cx,cy),15,(255,0,0),cv2.FILLED)
    
                mp_draw.draw_landmarks(image,hand_landmarks,mp_hands.HAND_CONNECTIONS)

    cv2.imshow('Hands_Detected',image)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
